````python
## Concatenating all the data:
data_list = []
dir_path = 'C:/Users/Acer/Documents/DSA4/ClassesNotebooks/DS4A_prg/cases/Cases_JupyterNote/case_1.2_student/datasets/'
dir_files = os.listdir(dir_path)
# Looping over the files
for file in dir_files:
    # Ths is to create the symbol later
    letter = file.split(".")[0]
    # Reading the data
    print("Reading the " ,file , " file")
    data_file = pd.read_csv(dir_path + file)
    # After reading the data, we have to create the new variables:
    # Symbol
    print("Creating all necessary features")
    data_file['Symbol'] = letter
    # All other features:
    data_file['Volume_Millions'] = data_file['Volume'] / 1000000.0
    data_file['VolStat'] = (data_file['High'] - data_file['Low']) /data_file['Open']
    data_file['Return'] = (data_file['Close'] /data_file['Open']) -1.0
    print("Appending the data ")
    data_list.append(data_file)
    print("Done, moving to the next file")
    print("\n")
    print("==================================")
data_total = pd.concat(data_list,axis = 0)
````

````python
label_list_data = []
for symbol in symbols:
    print("\n")
    print("Symbol: ", symbol)
    symbol_data = data_total[data_total['Symbol'] == symbol].copy()
    ## Tresholds
    treshold_symbol_1 = symbol_data.VolStat.quantile(0.25)
    treshold_symbol_2 = symbol_data.VolStat.quantile(0.75)
    max_symbol = symbol_data.VolStat.max()
    print("Quantile 0.25 : ", round(treshold_symbol_1,4))
    print("Quantile 0.75 : ", round(treshold_symbol_2,4))
    print("Max : ", round(max_symbol,4))
    
    
    # Feature Creation
    symbol_data['VolLevel'] = np.select(
        [
            symbol_data['VolStat'].between(0,treshold_symbol_1,inclusive = False),
            symbol_data['VolStat'].between(treshold_symbol_1,treshold_symbol_2,inclusive = True),
            symbol_data['VolStat'].between(treshold_symbol_2,max_symbol,inclusive = True)
        ],
        [
            'LOW',
            'MEDIUM',
            'HIGH'
        ]
    
    )
    print("Appending to the total data....")
    label_list_data.append(symbol_data)
    print(" ======================================== ")
final_labeled_data = pd.concat(label_list_data,axis = 0)
    
````
